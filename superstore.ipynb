```python
fig, ax = plt.subplots()

colors = {'Consumer':'Red', 'Corporate':'Blue', 'Home Office':'Green'}
ax.scatter(x=df['Sales'], y=df['Profit'], c=df['Segment'].apply(lambda x: colors[x])) # Scatter plot of different customer segments along with their sales and profit
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.title("Sales and Profit Analysis on all the segments")
plt.show()
```

##### as we can see the profit is more in consumer so lets analyse more about it to maximize profit

```python
df_sub_category = df.groupby(['Sub-Category'])[['Sales', 'Discount', 'Profit']].mean()
df_sub_category.head(10)
```

```python
plt.figure(figsize = (15,15))
plt.pie(df_sub_category['Sales'], labels = df_sub_category.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Sales Analysis', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

##### 1.copiers 2. Machines and so on have highest sales

# profit and discount relationship in consumer segment

```python
df2=df.loc[(df['Segment']=='Consumer')&(df['Discount']==0.1)] # If the discount is 10%
df2['Profit'].plot.hist(bins=50)
```

```python
df2=df.loc[(df['Segment']=='Consumer')&(df['Discount']==0.2)] # If the discount is 20%
df2['Profit'].plot.hist(bins=50)
```

```python
df2=df.loc[(df['Segment']=='Consumer')&(df['Discount']==0.5)] # If the discount is 50%
df2['Profit'].plot.hist(bins=50)
```

### it can be concluded that store owner makes more profit in 10% discount and it decreases as discount increases, so discounts more than 10% should be avoided mostly
